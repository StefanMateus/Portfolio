```python
D15 = pd.read_csv('../input/world-happiness/2015.csv')
D16 = pd.read_csv('../input/world-happiness/2016.csv')
D17 = pd.read_csv('../input/world-happiness/2017.csv')
D18 = pd.read_csv('../input/world-happiness/2018.csv')
D19 = pd.read_csv('../input/world-happiness/2019.csv')
```

```python
'''1. Display all top 10 countries in each year side by side in screen'''
#Just creating the 'vector' with the info we need to plot the graphics next
Top10_2015 = D15.loc[D15['Happiness Rank']<=10]
Top10_2016 = D16.loc[D16['Happiness Rank']<=10]
Top10_2017 = D17.loc[D17['Happiness.Rank']<=10]
Top10_2018 = D18.loc[D18['Overall rank']<=10]
Top10_2019 = D19.loc[D19['Overall rank']<=10]
```

```python
#Plotting the graphics side by side, just like a dashboard

#2015 Graphic
plt.subplot(2,3,1)
plt.bar(Top10_2015['Country'],Top10_2015['Happiness Score'])
plt.xticks(rotation = 45)
plt.title('Top 10 Happiest Countries in 2015')

#2016 Graphic
plt.subplot(2,3,2)
plt.bar(Top10_2016['Country'],Top10_2016['Happiness Score'])
plt.xticks(rotation = 45)
plt.title('Top 10 Happiest Countries in 2016')

#2017 Graphic
plt.subplot(2,3,3)
plt.bar(Top10_2017['Country'],Top10_2017['Happiness.Score'])
plt.xticks(rotation = 45)
plt.title('Top 10 Happiest Countries in 2017')

#2018 Graphic
plt.subplot(2,3,4)
plt.bar(Top10_2018['Country or region'],Top10_2018['Score'])
plt.xticks(rotation = 45)
plt.title('Top 10 Happiest Countries in 2018')

#2018 Graphic
plt.subplot(2,3,5)
plt.bar(Top10_2019['Country or region'],Top10_2019['Score'])
plt.xticks(rotation = 45)
plt.title('Top 10 Happiest Countries in 2019')

plt.tight_layout(rect=[0,0,3,2])
plt.show()
```

```python
'''2. Top 10 happiest countries'''
#JMerging the databases to build a timeseries with the top 10 happiest countries
HappyC = pd.DataFrame.merge(D15,D16,how = 'outer',on = 'Country',suffixes = ['_2015','_2016'])
HappyC = pd.DataFrame.merge(HappyC,D17,how = 'outer',on = 'Country',suffixes = ['','_2017'])
HappyC = pd.DataFrame.merge(HappyC,D18,how = 'outer',left_on = 'Country', right_on = 'Country or region',suffixes = ['','_2018'])
HappyC = pd.DataFrame.merge(HappyC,D19,how = 'outer',left_on = 'Country',right_on = 'Country or region' ,suffixes = ['','_2019'])
HappyC = HappyC.dropna(subset = ['Country'])
HappyC
```

```python
'''3. Which continent is the happier one?'''
Score = pd.DataFrame(HappyC['Country'])
Score.loc[:,'Region'] = HappyC['Region_2015']
Score.loc[:,'2015'] = (HappyC['Happiness Score_2015'])
Score.loc[:,'2016'] = HappyC['Happiness Score_2016']
Score.loc[:,'2017'] = HappyC['Happiness.Score']
Score.loc[:,'2018'] = HappyC['Score']
Score.loc[:,'2019'] = HappyC['Score_2019']
Score['Average'] = Score.mean(axis = 1) 
pd.to_numeric(Score['Average'])
Score
```

```python
Regions = Score.groupby('Region').mean()
Regions = Regions.sort_values(by = 'Average', ascending = False)
Regions

```

```python
plt.rcParams['xtick.labelsize'] = 8
plt.bar(Regions.index,Regions['Average'])
plt.xticks(rotation = 90)
plt.title('Happiest Regions in the World')
```

```python
#4. The Top 1 countries in each columns 
#Creating GDP data to see the best score in GDP category all over the years
GDP = pd.DataFrame(HappyC['Country'])
GDP.loc[:,'2015'] = HappyC['Economy (GDP per Capita)_2015']
GDP.loc[:,'2016'] = HappyC['Economy (GDP per Capita)_2016']
GDP.loc[:,'2017'] = HappyC['Economy..GDP.per.Capita.']
GDP.loc[:,'2018'] = HappyC['GDP per capita']
GDP.loc[:,'2019'] = HappyC['GDP per capita_2019']
GDP['Average'] = GDP.mean(axis = 1) 
pd.to_numeric(GDP['Average'])
GDP = GDP.sort_values(by = 'Average', ascending = False)

#Creating Family data to see the best score in Family category all over the years
Family = pd.DataFrame(HappyC['Country'])
Family.loc[:,'2015'] = HappyC['Family_2015']
Family.loc[:,'2016'] = HappyC['Family_2016']
Family.loc[:,'2017'] = HappyC['Family']
Family.loc[:,'2018'] = HappyC['Social support']
Family.loc[:,'2019'] = HappyC['Social support_2019']
Family['Average'] = Family.mean(axis = 1) 
pd.to_numeric(Family['Average'])
Family = Family.sort_values(by = 'Average', axis = 0,ascending = False)

#Creating Healthy data to see the best score in Healthy category all over the years
Healthy = pd.DataFrame(HappyC['Country'])
Healthy.loc[:,'2015'] = HappyC['Health (Life Expectancy)_2015']
Healthy.loc[:,'2016'] = HappyC['Health (Life Expectancy)_2016']
Healthy.loc[:,'2017'] = HappyC['Health..Life.Expectancy.']
Healthy.loc[:,'2018'] = HappyC['Healthy life expectancy']
Healthy.loc[:,'2019'] = HappyC['Healthy life expectancy_2019']
Healthy['Average'] = Healthy.mean(axis = 1) 
pd.to_numeric(Healthy['Average'])
Healthy = Healthy.sort_values(by = 'Average', ascending = False)

#Creating Freedom data to see the best score in Freedom category all over the years
Freedom = pd.DataFrame(HappyC['Country'])
Freedom.loc[:,'2015'] = HappyC['Freedom_2015']
Freedom.loc[:,'2016'] = HappyC['Freedom_2016']
Freedom.loc[:,'2017'] = HappyC['Freedom']
Freedom.loc[:,'2018'] = HappyC['Freedom to make life choices']
Freedom.loc[:,'2019'] = HappyC['Freedom to make life choices_2019']
Freedom['Average'] = Freedom.mean(axis = 1) 
pd.to_numeric(Freedom['Average'])
Freedom = Freedom.sort_values(by = 'Average', ascending = False)

#Creating Corruption data to see the best score in Corruption category all over the years
Corruption = pd.DataFrame(HappyC['Country'])
Corruption.loc[:,'2015'] = HappyC['Trust (Government Corruption)_2015']
Corruption.loc[:,'2016'] = HappyC['Trust (Government Corruption)_2016']
Corruption.loc[:,'2017'] = HappyC['Trust..Government.Corruption.']
Corruption.loc[:,'2018'] = HappyC['Perceptions of corruption']
Corruption.loc[:,'2019'] = HappyC['Perceptions of corruption_2019']
Corruption['Average'] = Corruption.mean(axis = 1) 
pd.to_numeric(Corruption['Average'])
Corruption = Corruption.sort_values(by = 'Average', axis = 0, ascending = True)


#Creating Generosity data to see the best score in Generosity category all over the years
Generosity = pd.DataFrame(HappyC['Country'])
Generosity.loc[:,'2015'] = HappyC['Generosity_2015']
Generosity.loc[:,'2016'] = HappyC['Generosity_2016']
Generosity.loc[:,'2017'] = HappyC['Generosity']
Generosity.loc[:,'2018'] = HappyC['Generosity_2018']
Generosity.loc[:,'2019'] = HappyC['Generosity_2019']
Generosity['Average'] = Generosity.mean(axis = 1) 
pd.to_numeric(Generosity['Average'])
Generosity = Generosity.sort_values(by = 'Average', axis = 0, ascending = False)
```
