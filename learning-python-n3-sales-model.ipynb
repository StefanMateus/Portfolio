```python
sales_data = pd.read_csv('../input/dummy-advertising-and-sales-data/Dummy Data HSS.csv')
sales_data.dropna()
```

```python
def multiple_graph(n_graficos):
    ind = np.arange(n_graficos)
    width = 0.15
    dist = []
    cont = 0
    while cont < n_graficos:
        dist.append(ind+(width*cont))
        cont += 1
    return dist

def autolabel(rects):
    for rect in rects:
        h = rect.get_height()
        ax.text(rect.get_x()+rect.get_width()/2., 1.05*h, '%d'%int(h),
                ha='center', va='bottom')

        
    
    
    
```

```python
#Understing what kind of data we are dealing with
influencer = sales_data.groupby('Influencer').sum()

width = 0.15
fig = plt.figure()
ax = fig.add_subplot(111)

distance = multiple_graph(4)
macro = ax.bar(distance[0], influencer['TV'],width = width)
micro = ax.bar(distance[1], influencer['Radio'], width = width)
mega = ax.bar(distance[2], influencer['Social Media'], width = width)
nano = ax.bar(distance[3], influencer['Sales'], width = width)

ax.set_xticks(distance[1])
ax.set_xticklabels(influencer.columns)
ax.legend(influencer.index)

autolabel(macro)
autolabel(micro)
autolabel(mega)
autolabel(nano)
```

```python
#Transforming the categorical data (Influencer) to a numerical data (Influencer_Categories)
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
cat_influencer = le.fit_transform(sales_data['Influencer'])
sales_data['Influencer_Categories'] = cat_influencer
sales_data
```

```python
#Understanding the correlation between each of the variables
import seaborn  as sn
corr_matrix = sales_data.corr()
sn.heatmap(corr_matrix,annot = True)
plt.title('Correlation Matrix')
plt.show()
```

```python
#Using linear regression to predict sales
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer

linear_model = LinearRegression(normalize = True)
my_imputer = SimpleImputer()
sales_data = sales_data.drop('Influencer', axis = 1)




```

```python
data_model = my_imputer.fit_transform(sales_data)
data_model = pd.DataFrame(data_model)
data_model.rename(columns = {0:'TV',1:'Radio',2:'Social Media',3:'Sales',4:'Influencer Categories'}, inplace = True)
```

```python
X = data_model.drop('Sales', axis = 1)
Y = data_model['Sales']

X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 0.3)

linear_model.fit(X_train,Y_train)
accuracy = linear_model.score(X_test,Y_test)

print(f'This Linear model accuracy is {accuracy}')
```
